```python
import pandas 

dataset = pandas.read_csv('50_Startups.csv')

dataset
```

```python
dataset =pandas.get_dummies(dataset,drop_first=True)
dataset
```

```python
dataset.columns
```

```python
indep = dataset[['R&D Spend', 'Administration', 'Marketing Spend',
       'State_Florida', 'State_New York']]

depe = dataset[[ 'Profit']]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(indep,depe,test_size=0.33,random_state=0)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(criterion='poisson',splitter="best")
regressor = regressor.fit(x_train,y_train)
```

```python
import matplotlib.pyplot as plt
from sklearn import tree
tree.plot_tree(regressor)
plt.show() 
```

```python
y_pred = regressor.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r_score = r2_score(y_test,y_pred)
```
